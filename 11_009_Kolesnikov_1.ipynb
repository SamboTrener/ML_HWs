```python
from matplotlib.image import imread
import numpy as np
import matplotlib.pyplot as plt

orig_img = imread('seles.jpg')

plt.imshow(orig_img)
plt.show()
```

```python
def mask(img, y_min, y_max, x_min, x_max, count):
    masked = np.asarray(img.copy())

    patch_width = (x_max - x_min) // count
    patch_height = (y_max - y_min) // count

    for x in range(count):
        for y in range(count):
            x0 = x_min + patch_width * x
            x1 = x0 + patch_width
            y0 = y_min + patch_height * y
            y1 = y0 + patch_height

            masked[y0:y1, x0:x1] = np.mean(masked[y0:y1, x0:x1], axis=(0, 1))

    return masked

masked = mask(orig_img, 290, 430, 400, 600, 20)
plt.imshow(masked)
plt.show()
```
